# Vector and Scalar Multiplication 

```python
#Modules

import matplotlib.pyplot as pl
import numpy as np
```

## Vector *  Scalar means it shows the elements of the vector scalar number of time

```python
vec = [3,4,5,6]
s=2

vec*s
```

## vec * s while vec=np.array(vec) then it miltiplies the element of the vector

```python
vec = np.array([3,4,5,6])
s=2

vec*s
```

## ploting vectors in graph

```python
vec2d= np.array([1,2])
s1=2
s2=5
s3=-1
```

```python
pl.plot([0,vec2d[0]],[0,vec2d[1]],'bs-', label='v' )
pl.plot([0,s1*vec2d[0]],[0,s1*vec2d[1]],'ro-',label='v*s1' )
pl.plot([0,s2*vec2d[0]],[0,s2*vec2d[1]],'kp-',label='v*s2' )
pl.plot([0,s3*vec2d[0]],[0,s3*vec2d[1]],'g*-',label='v*s3' )


# pl.axis('square')
pl.xlim([-4,4])
pl.ylim([-4,4])
pl.grid()
pl.legend()
pl.show()
```

## creating 2 vectors


```python
v1= np.array([4,7,8,5])
v2= np.array([7,1,2,3])
```

## plotting each vector


```python
pl.plot([0,v1[0]],[0,v1[1]],[0,v1[2]],[0,v1[3]],'ro-',label='v1')
pl.plot([0,v2[0]],[0,v2[1]],[0,v2[2]],[0,v2[3]],'bo-',label='v2')

pl.grid()
pl.legend()
pl.xlim(-10,10)
pl.ylim(-10,10)
pl.show()
```

## V1 + V2


```python
v3=v1+v2

pl.plot([0,v3[0]],[0,v3[1]],[0,v3[2]],[0,v3[3]],'ro-',label='v1')

pl.grid()
pl.legend()
pl.xlim(-10,10)
pl.ylim(-10,10)
pl.show()
```

## v1-v2

```python
v4=v1-v2

pl.plot([0,v4[0]],[0,v4[1]],[0,v4[2]],[0,v4[3]],'ro-',label='v1')

pl.grid()
pl.legend()
pl.xlim(-10,10)
pl.ylim(-10,10)
pl.show()
```

## v1 * 4 +v2/2

```python
v5=(v1*4)+(v2/2)

pl.plot([0,v5[0]],[0,v5[1]],[0,v5[2]],[0,v5[3]],'ro-',label='v1')

pl.grid()
pl.legend()
pl.xlim(0,50)
pl.ylim(0,50)
pl.show()
```
